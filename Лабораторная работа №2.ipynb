## Обработка пропущенных данных 

```python
sportsmen = '/Users/eva/Documents/Учеба/jupnote/athlete_events.csv'
df_2 = pd.read_csv(sportsmen, sep=",")
```

```python
df_2
```

```python
df_2.info()
```

```python
df_2["Medal"].fillna(0, inplace = True)
```

```python
df_2.info()
```

```python
df_2.dropna(subset=['Age'], inplace=True)
```

```python
df_2.info()
```

```python
num_cols = []
for col in df_2.columns:
    # Количество пустых значений 
    temp_null_count = df_2[df_2[col].isnull()].shape[0]
    dt = str(df_2[col].dtype)
    if temp_null_count>0 and (dt=='float64' or dt=='int64'):
        num_cols.append(col)
        temp_perc = round((temp_null_count / 271116) * 100.0, 2)
        print('Колонка {}. Тип данных {}. Количество пустых значений {}, {}%.'.format(col, dt, temp_null_count, temp_perc))
```

```python
# Фильтр по колонкам с пропущенными значениями
data_num = df_2[num_cols]
data_num
```

```python
for col in data_num:
    plt.hist(df_2[col], 50)
    plt.xlabel(col)
    plt.show()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.impute import MissingIndicator
```

```python
data_num_Weight = data_num[['Weight']]
data_num_Weight.head()
```

```python
indicator = MissingIndicator()
mask_missing_values_only = indicator.fit_transform(data_num_Weight)
mask_missing_values_only
```

```python
strategies=['mean', 'median', 'most_frequent']
```

```python
def test_num_impute(strategy_param):
    imp_num = SimpleImputer(strategy=strategy_param)
    data_num_imp = imp_num.fit_transform(data_num_Weight)
    return data_num_imp[mask_missing_values_only]
```

```python
strategies[2], test_num_impute(strategies[2])
```

```python
def test_num_impute_col(dataset, column, strategy_param):
    temp_data = dataset[[column]]
    
    indicator = MissingIndicator()
    mask_missing_values_only = indicator.fit_transform(temp_data)
    
    imp_num = SimpleImputer(strategy=strategy_param)
    data_num_imp = imp_num.fit_transform(temp_data)
    
    filled_data = data_num_imp[mask_missing_values_only]
    
    return column, strategy_param, filled_data.size, filled_data[0], filled_data[filled_data.size-1]
```

```python
test_num_impute_col(df_2, 'Height', strategies[0])
```
