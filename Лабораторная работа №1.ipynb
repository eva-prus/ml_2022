### Описание содержимого таблицы:
Class - номер культиватора; 
Alcohol - градус напитка; 
Malic Acid - яблочная кислота; 
Ash - все неорганические вещества, оставшиеся после воспламенения остатка, оставшегося от испарения вина; 
Alcalinity of Ash - щелочность золы; 
Magnesium - содержания магния; 
Total Phenols - Общее количество фенолов; 
Flavanoids - Флавоноиды; 
Nonflavanoid Phenols - Нефлаваноидные фенолы; 
Proanthocyanins - Проантоцианины; 
Colour Intensity - Интенсивность цвета; 
Hue - Оттенок; 
OD280/OD315 of diluted wines - OD280/OD315 разбавленных вин; 
Proline - Пролин.

```python
# Проверим наличие пустых значений
# Цикл по колонкам датасета
for col in df.columns:
    # Количество пустых значений - все значения заполнены
    temp_null_count = df[df[col].isnull()].shape[0]
    print('{} - {}'.format(col, temp_null_count))
```
